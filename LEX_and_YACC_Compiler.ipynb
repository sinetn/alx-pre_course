## Lex only

```python
#@title Writing Lex program
%%writefile program.l


%%
ask {printf("TOKEN: KEYWORD (ask) \n");}
say {printf("TOKEN: KEYWORD (say) \n");}
if {printf("TOKEN: KEYWORD (if) \n");}
else {printf("TOKEN: KEYWORD (else) \n");}
end {printf("TOKEN: KEYWORD (end) \n");}
repeat {printf("TOKEN: KEYWORD (repeat) \n");}
while {printf("TOKEN: KEYWORD (while) \n");}

int {printf("TOKEN: DATA TYPE (int) \n");}
real {printf("TOKEN: DATA TYPE (real) \n");}
string {printf("TOKEN: DATA TYPE (string) \n");}

[a-zA-Z]([a-zA-Z0-9])*  {printf("TOKEN: IDENTIFIER (%s) \n", yytext);}


[ \t\n] {;}
.   {printf("Lexical error: %s \n", yytext);}







%%

void main(int argc, char *argv[]){
    if(argc!=2){
        printf("Usage:\n\t./a.out <FILENAME>\n");
        exit(0);
    }

    yyin=fopen(argv[1],"r");
    yylex();

    fclose(yyin);

}

int yywrap(){
    return 1;
}
```

if you want to use at txt as an input

```python
%%writefile program.txt

# Declaration
a : int
b : int
sum : int
i : int

# Input
say "Enter a and b: "
ask a
ask b

sum <- a + b

# Output 1
if sum >= 0
  say "Sum is non-negative"
else
  say "Sum is negative"
end

# Output 2
i <- 0
repeat while i < sum
  say i
  i <- i + 1
end
```

```python
#@title Shell Execution (you can rewrite the commands as per your need, eg. if you want to include a file as an input)
%%shell

lex -l program.l
gcc lex.yy.c
./a.out program.txt
```

## Lex and Yacc combined

```python
#@title Writing YACC program
%%writefile program.y

%{
    #include<stdio.h>
    #include<stdlib.h>
%}
%token DIGIT LETTER UND NL
%%
stmt: variable NL {printf("Valid Identifier\n");exit(0);}
variable: LETTER alphanumeric;
alphanumeric: LETTER alphanumeric
            | DIGIT alphanumeric
            | UND alphanumeric
            | LETTER
            | DIGIT
            | UND;
%%

int yyerror(){
    printf("Invalid Identifier\n");
    exit(0);
}

void main(){
    printf("Enter the variable name: ");
    yyparse();
}
```

```python
#@title Writing Lex program
%%writefile program.l

%{
    #include "y.tab.h"
%}
%%
[a-zA-Z] {return LETTER;}
[0-9] {return DIGIT;}
[_] {return UND;}
\n {return NL;}
. {return yytext[0];}
%%
```
